```python
from os import listdir
from json import loads, dumps

content = []
for each in listdir('./gets'):
    if each.endswith('.txt'):
        print(each)
        content += loads(open(f'./gets/{each}', encoding='utf-8').read())
with open(SUBMISSION_DATA_PATH, 'w', encoding='utf-8') as f:
    f.write(dumps(content, ensure_ascii=False, indent=4))
```

## 특별상

```python
from json import loads
df = loads(open(SUBMISSION_DATA_PATH).read())
```

```python
from collections import defaultdict
submission = defaultdict(list)
languages = defaultdict(set)
once_submitted_at = {}
for cnt, usr, prob, res, mem, runtime, lang, length, at in df:
    submission[usr].append(int(at))
    languages[usr].add(lang)
    if usr in once_submitted_at:
        once_submitted_at[usr] = -1
    else:
        once_submitted_at[usr] = int(at)
submission
```

## 잠시 소란이 있었지만 대회를 다시 진행하겠습니다

```python
submission_gap = {}
_max = (None, -1)
for each in submission:
    now = submission[each]
    ln = len(now)
    if ln < 2:
        continue
    submission_gap[each] = max([now[i] - now[i - 1] for i in range(ln - 1)])
    if _max[1] < submission_gap[each]:
        _max = (each, submission_gap[each])

_max
```
